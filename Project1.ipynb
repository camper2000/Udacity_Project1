Function that creates dummy columns for fields containing multiple answers seperated by ';'  

```python
def create_dummy_df2(df, cat_cols):
    """Convert categorical columns into many dummy columns"""
    
    for col in cat_cols:
        df = pd.concat([df.drop(col,axis=1),pd.get_dummies(df[col],prefix=col,prefix_sep='_', drop_first=True, dummy_na=True)],axis=1)
    return df
```

```python
def create_dummy_df(df, cat_cols):
    """This is revised based on above funcion. It is for fields containing multiple answers seperated by '; """
    
    for col in cat_cols:
        df = pd.concat([df.drop(col,axis=1),df[col].str.get_dummies(sep=';').rename(lambda x: col + '_' + x, axis='columns')],axis=1)
    return df
```

create dummy fields and fix fields with correct data type

```python
df = create_dummy_df(df, ['UndergradMajor','EduOther','DevType','JobSat','DevEnviron','JobFactors','WorkPlan','WorkChallenge','CodeRev'])
df.head()
```

```python
df = df.astype({"LanguageWorkedWith": str, "LanguageDesireNextYear": str, "PlatformWorkedWith": str, "PlatformDesireNextYear": str, "MiscTechWorkedWith": str, "MiscTechDesireNextYear": str})
df['Language'] = df[['LanguageWorkedWith', 'LanguageDesireNextYear']].agg(';'.join, axis=1)
df['Platform'] = df[['PlatformWorkedWith', 'PlatformDesireNextYear']].agg(';'.join, axis=1)
df['MiscTech'] = df[['MiscTechWorkedWith', 'MiscTechDesireNextYear']].agg(';'.join, axis=1)
df = df.drop(['LanguageWorkedWith', 'LanguageDesireNextYear','PlatformWorkedWith', 'PlatformDesireNextYear','MiscTechWorkedWith', 'MiscTechDesireNextYear'],axis=1)
df = create_dummy_df(df, ['Language','Platform','MiscTech'])
df.head()
```
